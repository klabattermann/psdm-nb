```python
import time
import numpy as np
import pandas as pd
from datetime import datetime

pd.options.plotting.backend = 'holoviews'
```

```python
def read_logs(fnames, tref=0):
    data = []
    for ind, fn in fnames:
        with open(fn) as fp:
            for line in fp:
                tok = line.split()
                tunix = time.mktime(time.strptime(" ".join(tok[:2]), "%Y%m%d %H:%M:%S"))
                if tunix < tref:
                    continue
                rate = int(tok[2].split('=')[1])
                sizeMB = int(tok[3].split('=')[1])
                if len(tok) >= 6:
                    osts = list(map(int, tok[5].rstrip(',').split(',')))
                    nosts = len(osts)
                    ost0 = osts[0]
                else:
                    nosts=0
                    ost0 = -1
                data.append((ind, tunix, sizeMB, rate, nosts, ost0))
    return data
```

```python
fns = [(4, 'xrd_dtn04_rates_20200818.log'), (3, 'xrd_dtn03_rates_20200818.log')]
start = datetime(2020, 8, 16, 9)
print(start.timestamp())
data = read_logs(fns, start.timestamp())

df = pd.DataFrame(data, columns=['dtn', 't', 'size', 'rate', 'nosts', 'ost0'] )
df.head()
df.info()
df.describe()
```
